```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import pandas as pd

# Load data
data = pd.read_csv("clustering.csv")

# Select relevant columns
data = data.iloc[:,6:11]  # Make sure these columns are numeric

# Convert non-numeric values to NaN
data = data.apply(pd.to_numeric, errors='coerce')
print(data)
# Drop rows with missing values (NaN)
data = data.dropna()

# Extract features for plotting and clustering
x = data["ApplicantIncome"]
y = data["LoanAmount"]

# Plot the data points
plt.figure()
plt.scatter(x, y, marker="o", facecolors="none", edgecolors="k", s=30)
plt.xlabel("ApplicantIncome")
plt.ylabel("LoanAmount")
plt.title("Scatter plot of Applicant Income vs Loan Amount")
plt.xticks(())
plt.yticks(())
plt.show()

# Train KMeans model on the numeric data
model = KMeans(init="k-means++", n_clusters=4)

# Train the model 
model.fit(data)

# Use the model to predict clusters
results = model.predict(data)
print(results)

# Find centroids
centroids = model.cluster_centers_
print(centroids)

# Plot the clusters along with centroids
plt.figure()
plt.scatter(x, y, marker='o', facecolors='none', edgecolors='k', s=30, c=results, cmap='viridis')
plt.scatter(centroids[:, 0], centroids[:, 1], marker='o', s=200, color='r')
plt.xlabel("ApplicantIncome")
plt.ylabel("LoanAmount")
plt.title("KMeans Clustering - Centroids and Boundaries")
plt.show()
```

## # kmeans clustring (wholsale cutomer data)

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import metrics
from  sklearn.cluster import KMeans
import pandas as pd 

#step 1 lord the dataset
data = pd.read_csv("Wholesale customers data.csv")
# print(data.to_string())
data.info()

x=data["Fresh"]
y=data["Grocery"]
# y=data["Region"]
data=list(zip(x,y))

#plot the datasets to identify the clusteres
plt.figure()
plt.scatter(x,y,marker="o", facecolors="none",edgecolors="k",s=30)
plt.title('Input Data')
plt.xticks(())
plt.yticks(())

#lord the model
model=KMeans(init="k-means++",n_clusters=4)

#train the model 
model.fit(data)

#use the model
results=model.predict(data)
print(results)

#find centeroids
centroids=model.cluster_centers_
print(centroids)

#plot the clusters along with centroids
plt.figure()
plt.scatter(x,y,marker='o',facecolors='none',edgecolors='k',s=30)
plt.scatter(centroids[:,0],centroids[:,1],marker='o',s=200,color='r')
plt.title('centroids on boundries obtaind using KMeans') 
```

### k means clustring(2017 data)

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import metrics
from  sklearn.cluster import KMeans
import pandas as pd 

#step 1 lord the dataset
data = pd.read_csv("2017.csv")
# print(data.to_string())

x=data["Happiness.Rank"]
y=data["Happiness.Score"]
data=list(zip(x,y))


#plot the datasets to identify the clusteres
plt.figure()
plt.scatter(x,y,  marker="o", facecolors="none",edgecolors="k",s=30)
plt.title('Input Data')
plt.xticks(())
plt.yticks(())

#lord the model
model=KMeans(init="k-means++",n_clusters=4)

#train the model 
model.fit(data)

#use the model
results=model.predict(data)
print(results)

#find centeroids
centroids=model.cluster_centers_
print(centroids)

#plot the clusters along with centroids
plt.figure()
plt.scatter(x,y,marker='o',facecolors='none',edgecolors='k',s=30)
plt.scatter(centroids[:,0],centroids[:,1],marker='o',s=200,color='r')
plt.title('centroids on boundries obtaind using KMeans')  
```

#### Meanshift Clustering

```python
from sklearn.cluster import MeanShift, estimate_bandwidth
import pandas as pd 
#lord the dataset
data=pd.read_csv("2017.csv")

#preprocess
data=data.iloc[:,1:]
print(data)
print("=========================")
#calculate the bandwidth
bandwidth=estimate_bandwidth(data)
print(bandwidth)

print("====================================")
#load the model
meanshift=MeanShift(bandwidth=bandwidth)
print(meanshift)
print("==========================================")
#train the model
result=meanshift.fit(data)
print(result)
print("===================r============================")
#show the result
result.predict(data)
```

```python
from sklearn.cluster import MeanShift, estimate_bandwidth
import pandas as pd 

#lord the dataset
data=pd.read_csv("clustering.csv")

#preprocess
data=data.iloc[:,6:11]
print(data)

# Convert non-numeric values to NaN
data = data.apply(pd.to_numeric, errors='coerce')
print(data)

# Drop rows with missing values (NaN)
data = data.dropna()

#calculate the bandwidth
bandwidth=estimate_bandwidth(data)
print(bandwidth)

#load the model
meanshift=MeanShift(bandwidth=bandwidth)
print(meanshift)

#train the model
result=meanshift.fit(data)
print(result)

#show the result
result.predict(data)
```
