# Regression 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model

# load the data
data=pd.read_csv("Salary.csv")

# Split the dataset
x=np.array(data.YearsExperience).reshape(-1,1)
y=np.array(data.Salary).reshape(-1,1)

# load the model
linear_regression_model=linear_model.LinearRegression()

# Train the model :fit
linear_regression_model.fit(x,y)

# Test the model
xtest=np.array([3.0,1.0,9.0,5.2]).reshape(-1,1)

ytest_result=linear_regression_model.predict(xtest)
print(xtest)
print(y)
print(ytest_result)


# Test the model
xtest=x
ytest_result=linear_regression_model.predict(xtest)
# visulization the actual value v/s predicated value
plt.scatter(x,y,color="red",label="actual")
plt.scatter(x,ytest_result,color="blue",label="predict")
plt.xlabel("XTest")
plt.ylabel("YTest")
plt.legend()
plt.show()
```

# Linear regression on  Weather data 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn import linear_model,model_selection
import matplotlib.pyplot as plt

# load the dataset
data=pd.read_csv('Weather.csv')
# print(data)

# inplace=should you want to change original data frame or not
data.replace('#VALUE!',np.nan,inplace=True)
data.isnull().sum()
ndf=data[['MaxTemp','Snowfall']]
ndf.dropna(axis=0,inplace=True)
print(ndf.isnull().sum())

# identify x and y
x=np.array(ndf.MaxTemp).reshape(-1,1)
y=np.array(pd.to_numeric(ndf.Snowfall)).reshape(-1,1)
print(ndf.dtypes)
# split the dataset in traning and testing
xtrain,xtest,ytrain,ytest=model_selection.train_test_split(x,y,test_size=0.1)

# load the model
lr=linear_model.LinearRegression()

# train the model
lr.fit(xtrain,ytrain)

# test the model
ytest_predicted=lr.predict(xtest)
print(ytest)
print(ytest_predicted)

plt.scatter(xtrain,ytrain,color="red")
plt.scatter(xtest,ytest_predicted,color="blue")
plt.xlabel("XLABEL")
plt.ylabel("YTABLE")
plt.show()
```

```python
import pandas as pd
from sklearn import model_selection, linear_model
import numpy as np
import matplotlib.pyplot as plt



# Load the data with low_memory=False to avoid dtype warnings
data = pd.read_csv('Weather.csv', low_memory=False)

# Preview the data to ensure it is loaded correctly
# print(data.head())
# print(data.info())

# Check for missing values and handle them if necessary
# print(data.isnull().sum())

# Assuming MinTemp is in a column named 'MinTemp' and MaxTemp in 'MaxTemp'
# Ensure columns have the correct data types
data['MinTemp'] = pd.to_numeric(data['MinTemp'], errors='coerce')
data['MaxTemp'] = pd.to_numeric(data['MaxTemp'], errors='coerce')

# Drop rows with NaN values in 'MinTemp' and 'MaxTemp'
data.dropna(subset=['MinTemp', 'MaxTemp'], inplace=True)

# Verify the shape of the dataset
# print(data.shape)

# Prepare the feature matrix (X) and target vector (y)
X = data['MinTemp'].values.reshape(-1, 1)
y = data['MaxTemp'].values

# Split the dataset into training and testing sets
xtrain, xtest, ytrain, ytest = model_selection.train_test_split(X, y, test_size=0.3, random_state=42)

# Load the model
lr = linear_model.LinearRegression()

# Train the model
lr.fit(xtrain, ytrain)

# Predict using the model
predictions = lr.predict(xtest)

# # Print predictions and actual values for comparison
# print(predictions)
# print(ytest)

#PLOTING
plt.scatter(xtrain,ytrain,color="red")
plt.scatter(xtest,predictions,color="blue")
plt.show()
```
